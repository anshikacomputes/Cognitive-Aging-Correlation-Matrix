```
# =========================================================================
# Project 2: Cognitive Aging Correlation Data Visualizer
# Author: Anshika (BSc Zoology)
# GitHub: github.com/anshikacomputes/Cognitive-Aging-Correlation-Matrix
#
# Purpose: This project visualizes the correlation between age, gender, and
# different cognitive test scores (e.g., memory, focus, language), simulating
# the data collected in studies like SANSCOG.
# =========================================================================

# We will use Pandas for data handling and Seaborn/Matplotlib for powerful visualization.
# Biologists often use 'R' for this, but mastering it in Python is a massive technical edge.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# =========================================================================
# Step 1: Create a Simulated Dataset
# We use simulation to define the known relationships for clear demonstration.
# =========================================================================
data = {
    'Age': [25, 30, 45, 60, 75, 80, 22, 55, 65, 70],
    'Memory_Score': [95, 92, 85, 70, 50, 40, 98, 75, 60, 55],
    'Focus_Score': [90, 88, 80, 75, 65, 60, 92, 78, 68, 62],
    'Language_Score': [85, 85, 80, 80, 75, 70, 88, 78, 72, 68],
    'Gender': ['F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M'] # Binary for simplicity
}

# Load into a Pandas DataFrame - standard data structure in data science
df = pd.DataFrame(data)

# Step 2: Basic Data Diagnostics
print(f"--- Dataset Head ---")
print(df.head())
print(f"\n--- Statistical Summary ---")
print(df.describe())

# =========================================================================
# Step 3: Compute the Correlation Matrix
# We calculate the statistical correlation (R-squared) between all numerical variables.
# For simplicity, we drop the categorical 'Gender' column first.
# =========================================================================
numeric_df = df.drop('Gender', axis=1)
correlation_matrix = numeric_df.corr()

print(f"\n--- Computed Correlation Matrix ---")
print(correlation_matrix)

# =========================================================================
# Step 4: Generate a Heatmap Visualization
# This creates a publication-quality visual that is easy to interpret.
# A professor can understand this in 5 seconds.
# =========================================================================
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", vmin=-1, vmax=1)
plt.title('Correlation Heatmap: Age and Cognitive Scores')
plt.show() # In Colab, this shows the plot instantly.
```
